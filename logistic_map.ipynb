```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
```

```python
rate = 3.68
x = 0.5
xs = [x]

for i in range(500):
    x = rate * (x - x**2)
    xs.append(x)
    
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
ax.plot(xs, 'k.-', markersize=3, linewidth=0.1, alpha=0.6);
# ax.set_ylim(0, 1)
```

```python
xs = []
rates = []

for rate in np.arange(0, 4, 0.001):
    
    # initial value of x (population size)
    x = 0.5
    
    # iteate first 10000 times to stabilize
    for i in range(10000):
        x = rate * (x - x**2) # rate*x(1 - x)
    
    # continue iterating and save last (stable) vals
    stable = x
    for i in range(1000):
        x = rate * (x - x**2)
        rates.append(rate)
        xs.append(x)
        
        if abs(stable - x) < 0.0001:
            break
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
ax.plot(rates, xs, 'k,', alpha=0.3)
ax.set_xlim(3.4,4);
```
